```python
train_labels=pd.read_csv("G:\\competition\\kaggle\\invasive_species_monitoring\\train_labels.csv")
```

```python
#一次性将数据全部载入内存不够用，应该尝试分批载入训练数据;
#训练用图片共1500张，设定为每批次500，共三批；
def read_train_data(num):
    train_img = np.zeros(shape=(500, 150, 200, 3))
    train_label = np.array(train_labels.invasive.values[(0+500*num):(500+500*num)])
    for i in range(500):
        train_img[i]=transform.resize(io.imread(train_path+str(i+500*num+1)+'.jpg'),output_shape=(150,200,3))
#验证用图片为795张，同样分批次进行验证，共三批次，每批次265张图片;
def read_val_data(num):
    val_img = np.zeros(shape=(295, 150, 200, 3))
    val_label = np.array(train_labels.invasive.values[(1500+295*num):(1500+295+295*num)])
    for i in range(795):
        val_img[i]=transform(io.imread(train_path+str(i+1500+295*num+1)+'.jpg'),output_shape=(150,200,3))
```
